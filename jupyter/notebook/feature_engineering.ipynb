```python
scaler = MinMaxScaler()
scaler.fit(x_train[train_vars])

x_train = pd.concat([x_train[['id', 'smoking']].reset_index(drop=True), 
                     pd.DataFrame(scaler.transform(x_train[train_vars]), columns=train_vars)], 
                     axis = 1)

x_test = pd.concat([x_test[['id', 'smoking']].reset_index(drop = True), 
                     pd.DataFrame(scaler.transform(x_test[train_vars]), columns=train_vars)], 
                     axis = 1)
```
